## Ejercicio 1 

Dos dados se lanzan una vez y se observa el total obtenido. Usa una simulación para encontrar la probabilidad estimada de que la puntuación sea mayor a 7 o un número par. Una simulación es una repetición del mismo experimento multiples veces para observar su comportamiento:

- Ejecuta el experimento 1000 veces (lanza 2 dados 1000 veces, y suma el número de ambos dados).
- Lleva ua cuenta de los números y las veces que la suma fue mayor a 7 o un número par.
- Divide el número del paso 2 entre el número de iteraciones(1000).

```python
# TODO
import numpy as np
import random
n_iteraciones = 1000

def lanzamiento():
    return random.randint(1,6)

contador = 0
for i in range(n_iteraciones):
    lanz_1 = lanzamiento()
    lanz_2 = lanzamiento()
    suma = lanz_1+lanz_2
    if suma > 7 or suma%2 == 0:
        contador+=1

probabilidad = contador/n_iteraciones

print(f'La probabilidad experimental fue de {probabilidad}')
```

## Ejercicio 2

Una caja contiene 10 bolas blancas, 20 bolas rojas y 30 bolas verdes. Si tomamos 5 bolas de la caja reemplazándolas (tomando una bola, anotando el color y luego regresandola a la caja). Queremos saber la probabilidad de:

1. Tomar 3 blancas y 2 rojas.
2. Toas todas del mismo color.

Ejecuta el experimento 1000 veces y calcula las probabilidades que mencionadas.

```python
import random

ball_box = {}

# Crea la cajas con las bolas
for i in range(60):
    if i < 10:
        ball_box[i] = "White"
    elif (i > 9) and (i < 30):
        ball_box[i] = "Red"
    else:
        ball_box[i] = "Green"

print(ball_box)
            
# TODO
def sacar_bola():
    return ball_box[random.randint(0,59)]
print(sacar_bola())


def sacar5():
    bolas = []
    for i in range(5):
        bolas.append(sacar_bola())
    return bolas

def contar_colores(bolas):
    white = 0
    green = 0
    red = 0
    for i in range(len(bolas)):
        if bolas[i] == 'White':
            white +=1
        elif bolas[i] == 'Green':
            green +=1
        else:
            red+=1
    return {'White':white, 'Green':green, 'Red':red}
        
contador1 =0
contador2 = 0
for i in range(1000):
    dic = contar_colores(sacar5())
    if (dic['White']==3 and dic['Red']==2):
        contador1 +=1
    if(dic['White'] ==5 or dic['Red']==5 or dic['Green']==5):
        contador2 +=1

print(f'La probabilidad de sacar 3 blancas y 2 rojas es de {contador1/1000}\nLa probabilidad de sacar todas del mismo color es de {contador2/1000}')

```
